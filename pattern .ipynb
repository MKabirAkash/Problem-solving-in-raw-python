```python
import matplotlib.pyplot as plt
import numpy as np
import math



trials = 1000
hits = 0

hit_x = []
hit_y = []

miss_x = []
miss_y = []

for i in range(trials):
    x = np.random.uniform(low=-8, high=8)
    y = np.random.uniform(low=-6, high=6)
    dist=math.sqrt((x-3)**2+(y-1)**2)
    if(dist<=5):
      hits=hits+1
      hit_x.append(x)
      hit_y.append(y)
    else:
      miss_x.append(x)
      miss_y.append(y)


   

print(hits)
P_of_hit=hits/trials
PI=(P_of_hit*192)/25
print("value of pi: ")
print(PI)
plt.scatter(hit_x, hit_y, c='red')
plt.scatter(miss_x, miss_y, c='green')
```

```python
ax=[100]
ay=[100]
bx=[100]
by=[100]
ax[0]=0
ay[0]=10
x=0
y=10
for i in range(1,31):
  x=x+7
  y=y
  ax.append(x)
  ay.append(y)

bx[0]=20
by[0]=10
y1=10
for i in range(1,31):
  x1=20
  y1=y1-3
  bx.append(x1)
  by.append(y1)

import math 

cx = [100]
cy = [100]
cx[0]=40
cy[0]=0

S = 2


for t in range(0, 11):
    dist = math.sqrt((ax[t]-cx[t])**2+(ay[t]-cy[t])**2)

    cos_theta = (ax[t]-cx[t])/dist
    sin_theta = (ay[t]-cy[t])/dist

    new_px = cx[t] + S*cos_theta
    new_py = cy[t] + S*sin_theta
    
    cx.append(new_px)
    cy.append(new_py)
    print("Distance between A and  C in  second:  ",t)
    print("is :",dist)
for t in range(11, 21):
    dist1 = math.sqrt((bx[t]-cx[t])**2+(by[t]-cy[t])**2)
    

    cos_theta1 = (bx[t]-cx[t])/dist1
    sin_theta1 = (by[t]-cy[t])/dist1

    new_px1 = cx[t] + S*cos_theta1
    new_py1 = cy[t] + S*sin_theta1
    
    cx.append(new_px1)
    cy.append(new_py1)
    print("Distance between B and  C in  second:  ",t)
    print("is :",dist1)
for t in range(21, 31):
    dist2 = math.sqrt((ax[t]-cx[t])**2+(ay[t]-cy[t])**2)

    cos_theta2 = (ax[t]-cx[t])/dist2
    sin_theta2 = (ay[t]-cy[t])/dist2

    new_px2 = cx[t] + S*cos_theta2
    new_py2 = cy[t] + S*sin_theta2
    
    cx.append(new_px2)
    cy.append(new_py2)
    print("Distance between A and  C in  second:  ",t)
    print("is :",dist2)

```

```python
print("Coordinates of A")
for x,y in zip(ax, ay):
    print("(", x, ", " ,y, ")")
print("Coordinates of B")
for x,y in zip(bx, by):
    print("(", x, ", " ,y, ")")
print("Coordinates of C")
for x,y in zip(cx, cy):
    print("(", x, ", " ,y, ")")
print(len(ax))
```

```python
import matplotlib.pyplot as plt
import numpy as np

hits = 0
trials = 1000

hit_x = []
hit_y = []

miss_x = []
miss_y = []

for i in range(trials):
    x = np.random.uniform(low=-2, high=10)
    y = np.random.uniform(low=-2, high=8)

    # hit condition
    if x>=2 and x<=6:
        # rectangle check and circle check
        if y<=4:
            hits+=1
            hit_x.append(x)
            hit_y.append(y)
        elif y>4:
          if (x-4)**2 + (y-4)**2 <= 4:
            hit_x.append(x)
            hit_y.append(y)
          else:
            miss_x.append(x)
            miss_y.append(y)


    elif x>6 and x<=8:
      if y<=2:
        if  y>= x-6: 
      #triangle check
            hits+=1
            hit_x.append(x)
            hit_y.append(y)
        else:
           miss_x.append(x)
           miss_y.append(y)
      elif y>2:
        if y<=4 and y<(8-x+2):
            hits+=1
            hit_x.append(x)
            hit_y.append(y)
        else:
            miss_x.append(x)
            miss_y.append(y)
      else:
            miss_x.append(x)
            miss_y.append(y)
    else:
      miss_x.append(x)
      miss_y.append(y)

print(hits)
area = (hits/trials)*(12*10)
print(area)

plt.scatter(hit_x, hit_y, c='red')
plt.scatter(miss_x, miss_y, c='blue')
plt.show()
```

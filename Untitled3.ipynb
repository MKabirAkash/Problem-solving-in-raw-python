```python
import math
import numpy as np

n = 1000
f_sum = 0
f_sq_sum = 0
a = 0
b = 2

for i in range(n):
    x = np.random.uniform(low=a, high=b)
    y = 3*(x**3)+(x**2)+5

    f_sum += y
    f_sq_sum += (y**2)

f_avg = f_sum/n
f_sq_avg = f_sq_sum/n
integral = f_avg * (b-a)
error = (b-a)/math.sqrt(n)*math.sqrt(f_sq_avg - f_avg**2)
print("Intergral: " ,integral)
print("Error: " ,error)
```
